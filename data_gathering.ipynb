# Importing Dependancies

```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
import pymongo
import json
from pprint import pprint
from splinter import Browser
import time
import numpy as np
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.box_office_db
collection_box_office = db.box_office
collection_prod = db.numbers
```

```python
#https://stackoverflow.com/questions/20167194/insert-a-pandas-dataframe-into-mongodb-using-pymongo

#movie_title = []
for i in range(2015, 2021):
    #print(f'https://www.boxofficemojo.com/year/world/{i}')
    file_url = f'https://www.boxofficemojo.com/year/world/{i}'
    #movie_links.append(file_url)
    table = pd.read_html(file_url)
    table_data = table[0]
    table_data["Year"] = i
    table_data.columns = ['Rank', 'Title', 'Worldwide', 'Domestic', 
              'Domestic Percent', 'Foreign', 'Foreign Percent', 'Year']
    #movie_title.append(table_data["Title"])
    #print(type(table_data))
    #dfs.append(tables)
    df = table_data.to_json(orient='records')
    #dfs.append(df)
    records = json.loads(df)
    db.collection_box_office.insert_many(records)
```

```python
listings = db.collection_box_office.find()

for listing in listings:
    print(listing)
```

```python
box_office_df = pd.DataFrame(db.collection_box_office.find())
box_office_df.head()
#df_1.loc[df_1["Title"] == "Black Panther"]
#df_1.to_excel('export_1.xlsx')
```

```python
#omdb_data = []
for movie in box_office_df["Title"]:
    url_2 = 'http://www.omdbapi.com/?t='
    #api_key = "&apikey=545d5abd"
    api_key = "&apikey=trilogy"
    response = requests.get(f"{url_2}{movie}{api_key}")
    #print(response.url)
    data = response.json()
    #omdb_data.append(data)
    db.collection_box_office.update_many({"Title": movie}, {"$set": data}) 
```

```python
movie_df = pd.DataFrame(db.collection_box_office.find())
movie_df.head()
```

```python
array = np.arange(1, 5601, 100)
vars = array[1:]
vars
```

```python
url_list = []
url = 'https://www.the-numbers.com/movie/budgets/all'
url_list.append(url)

for var in vars:
    link = (f"{url}/{var}")
    url_list.append(link)
print(url_list)
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
for link in url_list:
    browser.visit(link)
    html = browser.html
    soup = bs(html, 'html.parser')
    table = soup.find('table')
    html_string = f"<html>{table}</html>"
    df = pd.read_html(html_string)
    df = df[0] 
    df = df[['Movie', 'ProductionBudget']]
    table = df.to_json(orient='records')
    records = json.loads(table)
    db.collection_prod.insert_many(records)
    time.sleep(6)
```

```python
prod_df = pd.DataFrame(db.collection_prod.find())
prod_df = prod_df.rename(columns={"Movie": "Title"})
prod_df.head()
```

```python
merged_df = pd.merge(movie_df, prod_df, on='Title')
merged_df
```
